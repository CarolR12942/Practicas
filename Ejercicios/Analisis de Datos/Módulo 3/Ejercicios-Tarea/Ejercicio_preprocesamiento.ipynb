# ⭕ Práctica: Dateset PIMA

The Pima Indians Diabetes Dataset involves predicting the onset of diabetes within 5 years in Pima Indians given medical details.

It is a binary (2-class) classification problem. The number of observations for each class is not balanced. There are 768 observations with 8 input variables and 1 output variable. **Missing values are believed to be encoded with zero values**. The variable names are as follows:

    Number of times pregnant.
    Plasma glucose concentration a 2 hours in an oral glucose tolerance test.
    Diastolic blood pressure (mm Hg).
    Triceps skinfold thickness (mm).
    2-Hour serum insulin (mu U/ml).
    Body mass index (weight in kg/(height in m)^2).
    Diabetes pedigree function.
    Age (years).
    Class variable (0 or 1).

Más información: https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

Quitamos los datos de la glucosa ya que son pocos datos

```python
df.dropna(subset=['glucosa'],inplace=True)#quita los valores faltantes
df.reset_index(inplace=True,drop=True)
display(df.isna().sum())
```

Con los demás datos faltantes, los imputamos ya que tienen una mayor relevacia en el diagnostico. Cabe resaltar que los datos de la variable de la glucosa se eliminaron para que no metieran ruido a las demás.

```python
from sklearn.impute import SimpleImputer

#para los embarazos, los imputamos por el "mas frecuente", ya que usar el promedio no tendria sentido
#ddecir que hay 3.5 embarazos
embarazos_values = df['embarazos'].values  # Recolectamos los valores de la columna en un arreglo de numpy

imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
#missin_values (cual es el valor faltante)np.nan, estrategia (se usa promedio)
#entonces con eso se "rellanara"
embarazos_new_values = imp.fit_transform(embarazos_values.reshape(-1,1))
#nuevas edades con el imp.fit
 #(-1,cuantos van aqui,1: necesita tener una column)
embarazos_new_values
```

```python
embarazos_new_values.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.title(f"Moda antes {round(np.nanmean(embarazos_values),5)}")
sns.histplot(embarazos_values)
plt.axvline(x=np.nanmean(embarazos_values))
plt.subplot(1,2,2)

sns.histplot(embarazos_new_values)
plt.title(f"Moda nuevo {round(np.mean(embarazos_new_values),5)}")
plt.axvline(x=np.mean(embarazos_new_values))
plt.show()
```

⚡ Hagamos binning (discretización):

```python
df_disc = df.copy()
df_disc['edad_grupos'] = pd.cut(df_disc['edad'],
                         bins=[ 12, 50, 99],
                      #0-12,12-50, 64-99
                         labels=[ 'adulto', 'tercera edad'])
#cut corta en intervalos
df_disc
```

```python
plt.figure()
sns.countplot(data=df_disc, x='edad_grupos', palette='viridis')

plt.title('Conteo de diagnostico por grupo de edad', fontsize=14)
plt.xlabel('Grupo de pacientes', fontsize=12)
plt.ylabel('Resultados', fontsize=12)

plt.tight_layout()
plt.show()
```
